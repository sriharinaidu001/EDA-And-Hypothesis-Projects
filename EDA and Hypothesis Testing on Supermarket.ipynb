# 1.Sample ttest or ttest_indepedence

```python
# This is test conducted between one catgorical and one numaric variable using groupby fucation
# Finding average/mean when 2groups/classes

# Null Hypothesis - No signficant deffrernce between variables/columns
# Alt. Hypothesis - Signficant deffrernce between variables/columns

# p-value less then 0.05 Reject null
# p-value grater then 0.05 Field to Reject null(Acceptance)
```

```python
from scipy.stats import ttest_ind
```

```python
# Find the average null hypothesis of Rating by Gender

supermarket.Rating.groupby(supermarket.Gender).mean()
```

```python
# Spliting Classes

GenderF=supermarket[supermarket.Gender=="Female"]
GenderM=supermarket[supermarket.Gender=="Male"]
```

```python
ttest_ind(GenderF.Rating,GenderM.Rating,equal_var=False)

# p-value is 0.5990065678416554 garter then 0.05 Failed to reject null
```

```python
# Find the average null hypothesis of Gross income by customer type

supermarket["gross income"].groupby(supermarket["Customer type"]).mean()
```

```python
# spliting classes

CustomerMember=supermarket[supermarket["Customer type"]=="Member"]
CustomerNormal=supermarket[supermarket["Customer type"]=="Normal"]
```

```python
ttest_ind(CustomerMember["gross income"],CustomerNormal["gross income"],equal_var=False)

# p-value is 0.031175361946825297 garter then 0.05 Failed to reject null
```

# 2.f_oneway / Anova

```python
# This is test conducted between one catgorical and one numaric variable using groupby fucation
# Finding average/mean when more then 2groups/classes

# Null Hypothesis - No signficant deffrernce between variables/columns
# Alt. Hypothesis - Signficant deffrernce between variables/columns

# p-value less then 0.05 Reject null
# p-value grater then 0.05 Field to Reject null(Acceptance)
```

```python
from scipy.stats import f_oneway
```

```python
# Find the average null hypothesis of gross income by city

supermarket["gross income"].groupby(supermarket.City).mean()
```

```python
# Spliting Data

cityM=supermarket[supermarket.City=="Mandalay"]
cityN=supermarket[supermarket.City=="Naypyitaw"]
cityY=supermarket[supermarket.City=="Yangon"]
```

```python
f_oneway(cityM["gross income"],cityN["gross income"],cityY["gross income"])

# p-value is 2.4062880140583703e-05 less then 0.05  reject null
```

```python
# Find the average null hypothesis of Rating by Product line

supermarket.Rating.groupby(supermarket["Product line"]).mean()
```

```python
# Spltting Data

productElectronic=supermarket[supermarket["Product line"]=="Electronic accessories"]
productFashion=supermarket[supermarket["Product line"]=="Fashion accessories"]
productFood=supermarket[supermarket["Product line"]=="Food and beverages"]
productHealth=supermarket[supermarket["Product line"]=="Health and beauty"]
productHome=supermarket[supermarket["Product line"]=="Home and lifestyle"]
productSports=supermarket[supermarket["Product line"]=="Sports and travel"]
```

```python
f_oneway(productElectronic.Rating,productFashion.Rating,productFood.Rating,productHealth.Rating,
        productHome.Rating,productSports.Rating)

# p-value is 4.658935890086712e-06 less then 0.05  reject null
```

# 3.chi2_contingency

```python
# This is test conducted between two catgorical variables using crosstab fucation

# Null Hypothesis - No relation/association  between variables/columns
# Alt. Hypothesis - relation/association  between variables/columns

# p-value less then 0.05 Reject null
# p-value grater then 0.05 Field to Reject null(Acceptance)
```

```python
from scipy.stats import chi2_contingency
```

```python
# Find null Association between gender and customer type

pd.crosstab(supermarket.Gender,supermarket["Customer type"])
```

```python
chi2_contingency(pd.crosstab(supermarket.Gender,supermarket["Customer type"]))

# p-value is 1.282309338298269e-05 less then 0.05  reject null
```

```python
# Find null Association between city and Product line

pd.crosstab(supermarket.City,supermarket["Product line"])
```

```python
chi2_contingency(pd.crosstab(supermarket.City,supermarket["Product line"]))

# p-value is 7.7328490330817395e-25 less then 0.05  reject null
```
