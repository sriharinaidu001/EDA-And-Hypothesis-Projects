## Group by

# Find out Top10 High Price Picked Players And Visualise with "BAR" Chart

```python
print (ipl.Auction_Price.groupby(ipl["Player Name"]).sum().nlargest(10))

# The groupby() function is used to create a groupby object, which can be used to group the data by one or more columns 
#  in a dataframe. Once you have created the groupby object, you can then apply a variety of aggregate functions to the groups, 
#     such as sum(), mean(), count(), min(), max(), median(), and more.
```

```python
df=ipl.Auction_Price.groupby(ipl["Player Name"]).sum().nlargest(10).plot(kind="bar",title="Top 10 Solded Player [High Price]",
                color=["Orange","Maroon","Tan","cyan","yellow","violet","grey","teal","red","olive"] ,               
                                figsize=(15,5))

for i in df.containers:
    df.bar_label(i,fontsize=15,label_type="center")
    plt.ylabel(" Auction Price in Lakh")
    plt.xlabel("Player Name")
```
